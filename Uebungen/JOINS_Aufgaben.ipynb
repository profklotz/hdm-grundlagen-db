**Vorbereitung**

Die Datenbank enthält eine minimalistische Grundstruktur eines sozialen
Netzwerks:

*   In der Tabelle user befinden sich alle Benutzer des sozialen Netzwerks. Benutzer
haben eine eindeutige Kennung (id), sowie eine E-Mailadresse und einen Anzeigenamen.
*   In der Tabelle email werden alle bekannten E-Mailadressen und der Zeitpunkt der
Anmeldung gespeichert. Aktuell besteht zwischen den Tabellen user und email
eine 1:1-Beziehung. Zukünftig ist jedoch angedacht, dass eine E-Mailadresse mehrere Benutzerkonten besitzen kann, daher wurden diese Informationen in separaten
Tabellen gespeichert.
*   Benutzer können einander folgen. Das bedeutet, jeder Benutzer kann beliebig vielen anderen Benutzern folgen. Diese Beziehung ist durch die Tabelle isfollowing
abgebildet. Die Tabelle hat lediglich zwei Attribute, eines für die Kennung des Benutzers, der folgt (Spalte followerId) und eines für die Kennung des Benutzers,
der verfolgt wird (Spalte followeeId).
*   Benutzer können Posts erstellen. Dies sind einfache Textnachrichten, die in der
Tabelle post gespeichert sind. Jede Nachricht hat eine Kennung (id), den Nachrichtentext und einen Autor, welcher im Attribut authorId gespeichert ist und eine
Fremdschlüsselbeziehung zur Tabelle user hat.
*    Benutzer können beliebig viele Nachrichten „liken“. Diese Beziehung ist in der Tabelle likes gespeichert, wo das Attribut postId auf den Post und userId auf den
Benutzer verweist.
