**Übung: Per Datenbanksprache mit einem DB-System interagieren**

*Ziel der Übung*

*   Wir werden gemeinsam eine sehr kleine, relationale Datenbank anlegen
*   Wir werden dafür per SQL mit der Datenbank interagieren
*   Wir werden DDL-Befehle verwenden, um Tabellen anzulegen und zu verändern
*   Wir werden DML-Befehle verwenden, um Datensätze zu erstellen, auszulesen, verändern und löschen
*  Hinweis: Es geht in dieser Übung nicht darum, ein Meister in SQL zu werden - wir werden SQL in den nachfolgenden
Abschnitten genauer behandeln












*Hinweis*

Es gibt eine Reihe von guten **Einführungen und Tutorials** in SQLite. Besonders empfehlenswert ist aus meiner Sicht die
Seite https://www.sqlitetutorial.net/. Hier können Sie weitere Details und Befehle zur Interaktion mit SQLite erhalten

Zudem ist es sehr hilfreich, während der Arbeit mit einer relationalen Datenbank eine **SQL Referenz** griffbereit zu haben, um
sich jederzeit eine genaue Beschreibung der Syntax der verschiedenen SQL Sprachelemente ansehen zu können. Da jedes
Datenbanksystem geringfügige Unterschiede in seiner Interpretation von SQL hat, ist es am besten, die jeweilige Referenz
des verwendeten Datenbanksystems zu verwenden – in unserem Fall also die von SQLite: https://www.sqlite.org/lang.html.

*Szenario*

Wir werden eine **kleine Datenbank** erstellen, um Kontakte zu verwalten. Dies machen wir direkt in einem DB-System
(SQLite), mit dem wir über die Datenbanksprache SQL interagieren. Zu erst werden wir die dafür notwendige Tabelle anlegen.
Dann werden wir alle CRUD-Operationen ausführen, d.h. neue Kontakte anlegen, diese auslesen, bestehende Kontakte
verändern und auch einen Kontakt löschen.
